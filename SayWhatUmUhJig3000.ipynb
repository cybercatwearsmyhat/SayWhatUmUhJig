### Import libaries

```python
import json
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
import plotly.express as px
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

### Set plot background style

```python
plt.style.use('dark_background')
```

### Load csv data to Dataframes

```python
transcriptSet_1_file = 'Data/TranscriptSet_1.csv'
transcriptSet_2_file = 'Data/TranscriptSet_2.csv'
transcriptSet_3_file = 'Data/TranscriptSet_3.csv'
transcriptSet_4_file = 'Data/TranscriptSet_4.csv'


transcriptSet_1 = pd.read_csv(transcriptSet_1_file)
transcriptSet_2 = pd.read_csv(transcriptSet_2_file)
transcriptSet_3 = pd.read_csv(transcriptSet_3_file)
transcriptSet_4 = pd.read_csv(transcriptSet_4_file)
```

### Define stopwords

#### Stopwords are not included in Word Cloud analysis

```python
script_stopwords = list(STOPWORDS)
```

## Example using Transcript_1

_Check Datatframe_

```python
transcriptSet_1.head(10)
```

### Create a <ins>Series</ins> from the Transcripts column of the DataFrame

```python
script_1 = transcriptSet_1['Transcript'].copy()
```

```python
script_1 = script_1.str.lower()
```

_Check Series_

```python
script_1.head()
```

### Join transcripts into a <ins>single string list</ins> to pass into the Word Cloud generator

```python
script_1_words = ' '.join(str(x) for x in script_1)
```

_Check String_

```python
#change list indices if needed
print(script_1_words[1000:2000])
```

### Load mask image. PNG format required.

```python
script_1_mask = np.array(Image.open("./Masks/mask_1.png"))
```

### Generate the Word Cloud

```python
#instantiate a word cloud object
#font_path is optional and requires a .ttc file
script_1_wc = WordCloud(max_words=150,stopwords=script_stopwords,colormap= 'spring', 
                        #font_path= 'cloudFont/Hannotate.ttc', 
                        collocations=False,
                        mask=script_1_mask, random_state=4, min_word_length=2)
# generate the word cloud
script_1_wc.generate(script_1_words)
```

### Display and save the Word Cloud

```python
plt.figure(figsize=(20, 20))
plt.imshow(script_1_wc, interpolation='bilinear')
plt.axis('off')
plt.savefig(f'Graphics/cloud_1',dpi = 300)
plt.show()
```
